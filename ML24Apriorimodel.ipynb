```python
# Apriory ML model---> apriori algorithm is an association rule learning algorithm that is used in unsupervides machine learning to find relationships between variables in
# large datasets. it is most commonly used for market basket anlysis understanding which items are often purchased together ex.--> if a customer buys bread and butter they
# are likely to buy milk too. the apriori algorithem help to find such rule automaticly from transaction data
# HOW APIRORI ALGORITHM WORKS--->
```

```python
# transition id         iten bought
# 1                     milk bread and butter
# 2                     bread butter and  gem
# 3                     milk bread butter gem
# 4                     milk and bread
```

```python
# step(1)-->generate item sets-->firstly frequently item sets (combination of items) that appear together frequently in transitions ex.> milk ,butter and butter gem
# step(2)--> calculate the support -> support tells how frequently an item set appers in the data .spport of a = no. of transaction containing / total transactions
# for item set milk and bread--> support=3/4--->75%
# step(3)--> filter by minimum support -> you define a minimum support trisold 0.5. only item sets with support >= 0.5 are considerd frequent item sets
# step(4)--> generate associate rules --> from the frequent item sets apriori generates rules like a=b,so if a cusfomer buys milk and bread they are likely to buy butter
# step(5)--> eveluate rules using confidence and lift--> confidence tells us how often rules the rule is true, [confidence of a to b = support of a union b / support of a]
# confidence of milk and bread--> butter ==> this is equal to support of milk,break,butter/support of milk and bread ,  support of milk bread and butter--> 2/4= 0.5
# support of milk and bread = 3/4= 0.75==> confidence ==>0.5/0.75== 0.67 , 67% of time when people buy milk and bread they also buy butter
# lift--> lift tells how much more likely the item b  is bought when a is bought ,compared to its general frequency life of a to b = confidence of a to b/support of b
# condition--> benifit--> if lift is > 1 (positive co-relation), if lift= 1 (no co-relation),if lift < 1 (negetive co-relation)
# ex.--> if support of butter is 0.75 then lift = 0.67/0.75=0.89---> since life is less then 1 ,so butter is slitly less likely to be bought with milk and bread
```

```python
# data preprocessing
transform=[]
for i in range(0,7501):
  transform.append([str(df.values[i,j]) for j in range(0,20)])

# training the dataset
from apyori import apriori
rules=apriori(transactions=transform,min_support=0.003,min_confidence=0.2,
              min_lift=3,min_length=2)
# visualizing the results
a=list(rules)
result=[list(a[i][0]) for i in range(0,len(a))]
result
```
